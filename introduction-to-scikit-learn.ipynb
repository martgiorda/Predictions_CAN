```python
###
### Fill missing value with Scikit
###

car_sales_missing = pd.read_csv('car-sales-extended-missing-data.csv')

car_sales_missing.dropna(subset=["Price"], inplace=True)

x = car_sales_missing.drop("Price",axis=1)
y = car_sales_missing["Price"]

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer = SimpleImputer(strategy="constant", fill_value=4)
num_imputer = SimpleImputer(strategy="mean")

#Define columns

cat_features=["Make", "Colour"]
door_features=["Doors"]
num_features=["Odometer (KM)"]

#Create an imputer
imputer = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_features),
    ("door_imputer", door_imputer, door_features),
    ("num_imputer", num_imputer, num_features),
])

# Transform
filled_x = imputer.fit_transform(x)

filled_x

car_sales_filled = pd.DataFrame(filled_x, columns=["Make","Colour","Doors","Odometer (KM)"])

car_sales_filled.isna().sum()

###
### Transform non-numerical into numerical
###

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make","Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([(
    "one_hot", 
    one_hot,
    categorical_features)],remainder="passthrough")

transformed_x = transformer.fit_transform(car_sales_filled)

###
### Fit model
###

np.random.seed(42)

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()

import pandas as pd
housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"])

housing_df["target"] = housing["target"]

### With Ridge

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
import numpy as np

## Import algorithm/estimator

np.random.seed(42)

x = housing_df.drop("target", axis=1)
y = housing_df["target"] #Median house price in 100k

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

#Instantiate and fit model on training set

model = Ridge()

#Fit with training data
model.fit(x_train,y_train)

#Score
model.score(x_test,y_test)
```

```python
### With SVM
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm

## Import algorithm/estimator

np.random.seed(42)

x = housing_df.drop("target", axis=1)
y = housing_df["target"] #Median house price in 100k

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

#Instantiate and fit model on training set

model = svm.SVR()

#Fit with training data
model.fit(x_train,y_train)

#Score
model.score(x_test,y_test)
```

```python
### With HistGradientBoostingClassifier

import numpy as np
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn import svm

## Import algorithm/estimator

np.random.seed(42)

x = housing_df.drop("target", axis=1)
y = housing_df["target"] #Median house price in 100k

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

#Instantiate and fit model on training set

model = HistGradientBoostingClassifier(max_iter=100)

#Fit with training data
model.fit(x_train,y_train)

#Score
model.score(x_test,y_test)
```

```python
### With LinearSVR

import numpy as np
from sklearn.svm import LinearSVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline


## Import algorithm/estimator

np.random.seed(42)

x = housing_df.drop("target", axis=1)
y = housing_df["target"] #Median house price in 100k

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

#Instantiate and fit model on training set

model = make_pipeline(StandardScaler(),
                     LinearSVR(random_state=0, tol=1e-5))

#Fit with training data
model.fit(x_train,y_train)

#Score
model.score(x_test,y_test)
```

```python
## Import the RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor

#Seed

np.random.seed(42)

#Data

x = housing_df.drop("target", axis=1)
y = housing_df["target"]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

model = RandomForestRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
